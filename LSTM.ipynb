```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input

```

```python
# Load and preprocess the dataset
df = pd.read_csv("C:/Users/Admin/Desktop/Test 4/demand_forecasting_data.csv")

df.head()
```

```python
# Convert the Date column to datetime format (adjust date column name if needed)
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

df
```

```python
# Aggregate the data by summing 'Demand' (or 'Base_Sales') per Date
df_daily = df.groupby('Date').agg({'Demand': 'sum'}).reset_index()

df_daily
```

```python
# Sort the data by Date
df_daily = df_daily.sort_values('Date')

df_daily
```

```python
# Prepare data for LSTM
data = df_daily['Demand'].values.reshape(-1, 1)

data
```

```python
# Normalize the data to range [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

scaled_data
```

```python
# Define training data size (80% for training)
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
test_data
```

```python
# Prepare the data for LSTM (create sequences of previous timesteps as features)
def create_dataset(dataset, time_step=10):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        X.append(dataset[i:(i+time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)


time_step = 10
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

# Print the outputs for training data
print("X_train (input sequences):")
print(X_train)

print("\nY_train (target values):")
print(y_train)

# Print the outputs for testing data
print("\nX_test (input sequences):")
print(X_test)

print("\nY_test (target values):")
print(y_test)
```

```python
# Reshape input to be [samples, time steps, features] which is the LSTM input format
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

X_train
X_test
```

```python
# Build the LSTM model
model = Sequential()
model.add(Input(shape=(time_step, 1)))  # Define the input shape here
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
# Inverse transform predictions and actual values
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = scaler.inverse_transform(y_pred)

y_test_inv
y_pred_inv
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
print(f'Root Mean Squared Error: {rmse}')

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
print(f'Root Mean Squared Error: {rmse}')
```

```python
# Plot results
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test_inv)), y_test_inv, color='blue', label='Actual')
plt.plot(range(len(y_pred_inv)), y_pred_inv, color='red', linestyle='--', label='Predicted')
plt.title('LSTM Model Predictions')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
