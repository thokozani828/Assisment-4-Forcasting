```python
 # Import the pandas library
import pandas as pd 
from prophet import Prophet
import warnings
from prophet.plot import plot_plotly


# Load the dataset
df = pd.read_csv("C:/Users/Admin/Desktop/Test 4/Historical Product Demand.csv")

# Display the first few rows of the dataset
df.head()
```

##APPLYING DATA PREPROCCESSING

```python
# Convert the 'Order Date' column to datetime format
df ['Date'] = pd.to_datetime (df['Date'])
df
```

```python
category_filter = 'Category_028'
frame = df[df['Product_Category'] == category_filter]

# Display only 'Date' and 'Order_Demand' columns
frame = frame[['Date', 'Order_Demand']]

# Rename columns to fit Prophet's requirements
frame.columns = ['ds', 'y']

# Convert the 'y' column to numeric, setting errors='coerce' will turn any non-numeric values into NaN
frame['y'] = pd.to_numeric(frame['y'], errors='coerce')

# Drop rows with NaN values, as Prophet cannot handle them
frame = frame.dropna()

print(frame.head())
```

##CREATING THE TIME SERIA MODEL

```python
# Initialize the Prophet model and fit it to the data
m = Prophet(interval_width=0.95)
training_run = m.fit(frame)
```

##EVALUATION OF THE MORDEL

```python

# Make sure 'future' is defined
future = m.make_future_dataframe(periods=30, freq='D')  # Create a DataFrame for the next 30 days

# Generate the forecast
forecast = m.predict(future)
forecast.tail()
```

```python
warnings.filterwarnings("ignore")

# Plot the forecast using Plotly
plot1 = m.plot(forecast)
```

```python
plot2 = m.plot_components(forecast)
```
